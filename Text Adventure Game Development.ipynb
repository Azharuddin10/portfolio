This text-based adventure game, The Quest of Mahishmati, transports you to the grand setting of the Baahubali movie series. Take on the role of a brave warrior in the historic Mahishmati kingdom, solving puzzles and taking on difficult tasks.

## The Story
The kingdom of Mahishmati, where myths and legends influence people's lives, is in disarray after the unsolved death of its adored king, Baahubali. With his brother, Bhallaladeva, in charge, the kingdom longs for its real hero. In your quest to learn the truth and fulfill your destiny, you will travel through thick forests, the royal palace, and the center of the kingdom's troubles.

## Game Objective
There are three key phases to your quest, and each one calls for bravery, discernment, and guile:

### The Forest of Secrets: Look around Mahishmati's mysterious forest to find buried clues and relics.
### The Chambers of Strategy: Discover the intricate network of deceit within the royal residence and win over important allies.
### The Arena of Valor: Show off your abilities in a series of tests that will try your fighting prowess and tactical acumen and ultimately lead you to the throne.

## Specific instructions for Playing
Make crucial decisions and navigate the game through a series of interactive prompts. Your choices will have a big effect on Mahishmati's destiny and the plot. Accept the mythology and folklore of Baahubali to direct you on the road to triumph. Features An adventure with a lot of narrative set in the fascinating world of Baahubali. making choices at pivotal points that influence your path. puzzles and challenges that put your knowledge of the epic saga to the test. Set out to bring justice and peace back to Mahishmati. This is where your journey starts.

```python
import time

# user's name
user_name = input('Enter your name: ')

# Welcome the user using the entered name
print(f'Welcome, {user_name}!')

# before starting game count 3
count = 3
while count > 0:
    print(count)
    count = count - 1
    input(prompt = "< Press ENTER to start. >\n")
    
# Start
def win():
    print("\nCongratulations! You have united Mahishmati and brought peace to the land.")

def fail():
    print("\nYour journey ends here. The future of Mahishmati remains uncertain.")

def arena_of_valor():
    print("\nStage 4: The Arena of Valor")
    print("You must now face a trial of combat.")
    weapon = input("Choose your weapon: sword or bow? (sword/bow) ")

    if weapon.lower() == "sword":
        print("Armed with a sword, you display great bravery and skill.")
        win()
    elif weapon.lower() == "bow":
        print("The bow was a poor choice for close combat.")
        fail()
    else:
        print("Invalid choice. Please choose 'sword' or 'bow'.")
        arena_of_valor()

def chambers_of_strategy():
    print("\nStage 3: The Chambers of Strategy")
    print("You need to gain the trust of a key ally.")
    ally = input("Who will you approach: the minister, the general, or the king? (minister/general/king) ")

    if ally.lower() == "minister":
        print("The minister offers you wise counsel.")
        chambers_of_fight()
    elif ally.lower() == "general":
        print("The general suspects your motives and turns you away.")
        fail()
    elif ally.lower() == "king":
        print("The king welcomes you and pledges support to your cause.")
        chambers_of_fight()
    else:
        print("Invalid choice. Please choose 'minister', 'general', or 'king'.")
        chambers_of_strategy()

def forest_of_secrets():
    print("\nStage 1: The Forest of Secrets")
    attempts = 0
    max_attempts = 7

    while attempts < max_attempts:
        direction = input("You enter the forest. Which way do you go: left, right, or straight? (left/right/straight) ")
        if direction.lower() == "left":
            print("You find a hidden path that leads to the palace.")
            chambers_of_strategy()
            break
        elif direction.lower() == "right":
            attempts += 1
            print(f"You encounter a pack of wolves. You have {max_attempts - attempts} attempts left.")
        elif direction.lower() == "straight":
            print("You stumble upon a magical artifact that enhances your abilities.")
            chambers_of_fight()
            break
        else:
            attempts += 1
            print(f"Invalid choice. You have {max_attempts - attempts} attempts left.")

def chambers_of_fight():
    print("\nStage 2: The Chambers of Fight")
    print("You face a series of one-on-one battles to prove your combat skills.")
    outcome = input("Do you choose to fight (yes/no)? ").lower()

    if outcome == "yes":
        print("You emerge victorious in the Chambers of Fight.")
        arena_of_valor()
    elif outcome == "no":
        print("You decide to avoid the fights and leave the chambers.")
        fail()
    else:
        print("Invalid choice. Please choose 'yes' or 'no'.")
        chambers_of_fight()

def throne_room():
    print("\nStage 5: The Throne Room")
    print("You have reached the final stage, where you confront the ruler of Mahishmati.")
    decision = input("Will you negotiate or challenge the ruler? (negotiate/challenge) ").lower()

    if decision == "negotiate":
        print("Your diplomatic skills shine, and a peaceful resolution is achieved.")
        win()
    elif decision == "challenge":
        print("You engage in a fierce battle with the ruler for the throne.")
        win()
    else:
        print("Your indecision leads to chaos, and Mahishmati falls into unrest.")
        fail()

def additional_stage_1():
    print("\nStage 6: The Hidden Temple")
    print("You discover a secret temple with ancient knowledge.")
    decision = input("Will you explore the temple or continue your journey? (explore/continue) ").lower()

    if decision == "explore":
        print("You gain valuable insights from the ancient texts.")
        additional_stage_2()
    elif decision == "continue":
        print("You decide to press on with your quest.")
        fail()
    else:
        print("Invalid choice. Please choose 'explore' or 'continue'.")
        additional_stage_1()

def additional_stage_2():
    print("\nStage 7: The Mystic Caves")
    print("You navigate through mysterious caves filled with challenges.")
    decision = input("Will you use stealth or confront the challenges head-on? (stealth/confront) ").lower()

    if decision == "stealth":
        print("Your stealthy approach helps you overcome the challenges.")
        additional_stage_3()
    elif decision == "confront":
        print("You face the challenges head-on, testing your skills.")
        fail()
    else:
        print("Invalid choice. Please choose 'stealth' or 'confront'.")
        additional_stage_2()

def additional_stage_3():
    print("\nStage 8: The Enchanted Lake")
    print("You reach a magical lake with hidden powers.")
    decision = input("Will you harness the lake's power or avoid its mysteries? (harness/avoid) ").lower()

    if decision == "harness":
        print("You harness the lake's power, gaining a mystical advantage.")
        win()
    elif decision == "avoid":
        print("You decide to avoid the mysteries, continuing your journey.")
        fail()
    else:
        print("Invalid choice. Please choose 'harness' or 'avoid'.")
        additional_stage_3()

def start_game():
    print("Welcome to The Quest of Mahishmati!")
    print("Your mission is to find the truth behind Baahubali's disappearance and unite the kingdom.")
    print("You have 3 minutes to complete the quest.")

    start_time = time.time()  # Game start time

    forest_of_secrets()  # Call existing game function

    end_time = time.time()  # Game end time
    duration = end_time - start_time  # Calculating the duration in seconds

    print(f"\nGame Duration: {duration:.2f} seconds")

    if duration > 180:
        print("Time's up! You took too long to complete the quest.")
    else:
        print("Congratulations on completing the quest!")    

# Start the game
start_game()
```
